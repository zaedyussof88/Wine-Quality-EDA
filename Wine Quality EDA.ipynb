```python
#Exploring distributions of numeric columns using histograms

import warnings
warnings.simplefilter(action = 'ignore', category = FutureWarning)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style = "whitegrid")

fig, axes = plt.subplots(nrows = 3, ncols = 4, figsize = (15, 10))

axes = axes.flatten()

numeric_columns = data.select_dtypes(include = ['float64']).columns

for i, column in enumerate(numeric_columns):
    sns.histplot(data[column], ax=axes[i], kde = True, bins = 20, color = 'darkred')
    axes[i].set_title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
#Scatter plot between fixed acidity and volatile acidity

plt.figure(figsize = (8, 6))
sns.scatterplot(x = 'fixed acidity', y = 'volatile acidity', data = data, color = 'darkred')
plt.title('Scatter Plot of Fixed Acidity vs Volatile Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Volatile Acidity')
plt.show()
```

```python
#Scatter plot between citric acid and residual sugar

plt.figure(figsize = (8, 6))
sns.scatterplot(x = 'citric acid', y = 'residual sugar', data = data, color='darkblue')
plt.title('Scatter Plot of citric acid and residual sugar')
plt.xlabel('citric acid')
plt.ylabel('residual sugar')
plt.show()
```

```python
#Scatter plot between chlorides and free sulpher dioxide

plt.figure(figsize = (8, 6))
sns.scatterplot(x = 'chlorides', y = 'free sulfur dioxide', data = data, color = 'darkgreen')
plt.title('Scatter Plot of chloride and free sulphure dioxide')
plt.xlabel('chlorides')
plt.ylabel('free sulfur dioxide')
plt.show()
```

```python
#Scatter plot between total sulphure dioxide and density

plt.figure(figsize = (8, 6))
sns.scatterplot(x = 'total sulfur dioxide', y='density', data = data, color = 'darkorange')
plt.title('Scatter Plot of total surface dioxide and  density ')
plt.xlabel('total sulfur dioxide ')
plt.ylabel('density')
plt.show()
```

```python
#Scatter plot between pH and sulphate

plt.figure(figsize = (8, 6))
sns.scatterplot(x = 'pH', y = 'sulphates', data = data, color = 'purple')
plt.title('Scatter Plot of pH and sulphate')
plt.xlabel('pH ')
plt.ylabel('sulphates')
plt.show()
```

```python
data = pd.read_csv('WineQT.csv')
data = data.iloc[:, :-3]
print(data)
```

```python
data.describe()
```

```python
data.info()
```

```python
print(data.columns)
```

```python
#Plot Histogram for 'fixed acidity' distribution

plt.figure(figsize = (8, 6))
sns.histplot(data['fixed acidity'], bins = 20, kde = True, color = 'darkred', edgecolor = 'black')
plt.title('Distribution of Fixed Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Frequency')
plt.show()
```

```python
#Plot Scatter plot for 'density' vs 'pH'
plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'density', y = 'pH', data = data, color = 'darkred')
plt.title('Relationship between Density and pH')
plt.xlabel('Density')
plt.ylabel('pH')
plt.show()
```

```python
#Plot Boxplot for 'total sulfur dioxide'

plt.figure(figsize = (8, 6))
sns.boxplot(x = data['total sulfur dioxide'])
plt.title('Distribution of Total Sulfur Dioxide')
plt.xlabel('Total Sulfur Dioxide')
plt.show()
```

```python
#Plot scatter plot for volatile acid and citric acid 
plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'volatile acidity', y = 'citric acid', data = data, color = 'darkblue')
plt.title('Relationship between volatile acidity and citric acid')
plt.xlabel('volatile acidity')
plt.ylabel('citric acid')
plt.show()
```

```python
#Created new columns total_acidity and sulfur_ratio

data['total_acidity'] = data['fixed acidity'] + data['volatile acidity']
data['sulfur_ratio'] = data['free sulfur dioxide'] / data['total sulfur dioxide']
data.head()
```

```python
#Merge Multiple Datasets

data2 = data.copy()
merged_data = pd.merge(data, data2, left_index = True, right_index = True, how = 'inner')
merged_data.head()
```

```python
#Grouping and Aggregation
# Grouping by 'pH' and calculating mean and sum for 'density and 'sulphates'

grouped_data = data.groupby('pH').agg({'density': 'mean', 'sulphates': 'sum'}).reset_index()
grouped_data.head()
```
